EXPLORATORY DATA ANALYSIS

Çok kullanılan metodlardan birisi olan, value_counts() metodu bir sütun içerisindeki her bir veriden kaç adet bulunduğu bilgisini verir. 

```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
```

```python
print(data['Type 1'].value_counts(dropna =False))
```

Çıktıyı inceleyelim. "Type 1" sütunu pokemonun türü hakkında bilgiler içerir. Mesela çıktının ilk satırında, 800 adet pokemondan 112 tanesinin su pokemonu olduğunu ifade eder.

İstatistikle ilgili bir kavram olan "outliers" ifadesinden bahsedelim. Bu kavram "ayrık, aykırı" anlamına gelmektedir.

Bir örnek üzerinden anlatalım.

1,4,5,6,8,9,11,12,13,14,15,16,17 şeklinde bir seri olsun. Median değeri serinin ortasındaki sayıdır. Serinin ortasındaki sayı 11 dir ve bu durumda median=11 dir. median kavramı "quartile 50%" şeklindede adlandırılıyor.

"lower quartile" kavramından bahsedelim. lower quartile, 1 ile 11 arasındaki median değeridir. Yani 1 ile 11 sayılarının ortasındaki sayıdır ve bu sayı 6'dır. lower quartile sayısı 6'dır. "Lower quartile" kavramına "Q1" de denir. Aynı zamanda "25% quartile" da denir. Yani "lower quartile" = Q1 = 25% quartile dır.

Birde "upper quartile" kavramından bahsedelim. 11 ile 17 arasındaki sayı upper quartile değeridir. Bu durumda upper quartile=14 dür. "upper quartile" kavramına "Q3" de denir. Aynı zamanda "75% quartile" da denir. 

Yani özetle:

Minimum ------------ Lower Quartile(Q1) ------------ Median ------------ Upper Quartile(Q3) ------------ Maximum      

Sayıların ortasındaki Median'dır. Median ile en sağdaki sayının ortasındaki değer Upper Quartile 'dir. Median ile en soldaki sayının ortasındaki değer Lower Quartile 'dir. 

Outlier: Bir istatistiğe göre bir data'daki diğer data'lardan aykırı olan yani çok yüksek ya da çok düşük olan data'dır. 

Mesela 1000 tane ev olsun. Bu evlerden en ucuzu 100 dolar, en pahalısı 1000 dolardır. 1001. evin fiyatını 1.000.000.000 dolar olarak belirlersek bu bir outlier'dir. Çünkü diğer data'lardan çok yukarıda bir değer. Bu yüzden bu bir aykırı sayıdır. 

Peki ne kadar yukarıda olduğu zaman aykırı sayı olarak nitelendirebiliriz?:

Q3 ile Q1 arasındaki mesafeyi "IQR" olarak adlandırırız. Yani IQR=Q3-Q1 dir. IQR değerini 1.5 ile çarpıp, Q1 den çıkarırız ve elde edilen değerin altındakilere outlier deriz. Yani "Q1 - 1.5(Q3-Q1)" ifadesinden elde edilen değerin altındaki değerlere outlier deriz. 1.5*IQR değeri Lower Quartile ile Minimum arasındaki mesafedir. Q1 den 1.5*IQR değerini çıkarırsak Mimimum değerinin altındaki bir noktayı elde ederiz. Elde edilen bu nokta outlier değeridir. 

Veya Q3 ile Maximum arasındaki mesafe 1.5*IQR değeridir. Q3+1.5*IQR ifadesinin sonucunda Maximum değerinin üstündeki bir noktayı elde ederiz. Elde edilen bu nokta outlier değeridir. 

Yani "Q1 - 1.5(Q3-Q1)" ifadesinden elde edilen değerin altındaki değerler veya "Q3+1.5*IQR" fadesinden elde edilen değerin üstündeki değerler outlier değeridir.
