INDEXING PANDAS TIME SERIES

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("pokemon.csv")
```

```python
data.head()
```

Yukarıda görüldüğü üzere, indeks değerleri 0,1,2,3,4 şeklindedir. Pandas Time Serie'lerin özellikleri, indekslerin time serie'lerden oluşmasıdır. 

int, float, str gibi veri tiplerimiz olduğunu hatırlayalım. Bir başka veri tipi ise, 'datetime' dır. datetime = object 'dir.

parse_dates(boolean) şeklinde bir metod vardır. Bu metod yyyy-mm-dd formatındaki stringi yyyy-mm-dd hh:mm:ss dormatına çevirir.

```python
time_list = ["1992-03-08","1992-04-12"]
print(type(time_list[1])) 
datetime_object = pd.to_datetime(time_list)
print(type(datetime_object))
```

İki adet string tutan time_list isimli bir liste oluşturduk. Bu listenin 1.indeksindeki değerin tipinin string olduğunu çıktıda görüyoruz. 

Sonrasında listeyi pd.to_datetime ile datetime tipine dönüştürüyoruz. Ve çıktıda dönüşüm sonucunu görüyoruz.

```python
import warnings
warnings.filterwarnings("ignore")
#Yukarıda yaptığım şey, anlamını çözemediğim bir hatayı kapatmaktır.

data2 = data.head()
date_list = ["1992-01-10","1992-02-10","1992-03-10","1993-03-15","1993-03-16"]
datetime_object = pd.to_datetime(date_list)
data2["date"] = datetime_object

data2= data2.set_index("date")
data2 
```

Pokemon datamıza time serisi ile ilgili yapabileceğimiz bir zaman indeksi ya da zaman feature'si olmadığı için onun ilk 5 sample'ına bir time listesi ekledik.

Sonra bu time listesini pandas'ın to_datetime() metodu ile datatime objesine çevirdik ve data'mıza sütun ismi date olacak şekilde bu objeyi ekledik.

Hatırlarsanız indeksler 0,1,2,3,4 şeklindeydi ve burada set_index() metodu ile indekslerin date sütunundaki değerler olmasını sağladık. Böylece artık indeksler datetime objesidir. Yani data'mız artık bir time series data'dır.

Time series data, zamana bağlı data demektir.

Şimdi Time series data'mız ile yapabileceklerimizi görelim.

```python
print(data2.loc["1993-03-16"])
print(data2.loc["1992-03-10":"1993-03-16"])
```

loc ifadesi data'nın içinden belli bir indeksteki şeyleri çekmemizi sağlar. ama bizim indeksimiz artık time serisi olduğu için biz buna istediğimiz tarihi yazıp print ettirebiliyoruz.

Çıktıda görüldüğü üzere, indeksi 1993-03-16 olan satırın bilgilerini yazdırdık.

Birde iki time series arasını yazdırabilmemiz de mümkün. Burada 1992-03-10 ile 1993-03-16 indekleri arasındaki satırları yazdırdık.

RESAMPLING PANDAS TIME SERIES

Resampling: Bir data'nın içerisinden elde ettiğimiz yeni sample'lere denir.

Time series objesinin kullanışlı metodları vardır. Mesela "M" harfini kullanarak tüm ayları, istediğimiz ayları resample edebiliriz. Ya da "A" harfini kullanarak tüm yılları, istediğimiz yılları resample edebiliriz.

Mesela yılları resample etmek istiyoruz ve yılların ortalama hız değerlerini bulmak istiyoruz. 1992 yılından 3 tane time serisinden sample var. 1993 ten de 2 tane var. Hızı (Speed) bulmak için, yıla göre resample yaparız ve bunun hızının ortalamasını buluruz. Bunu time series ile yapmak çok kolay çünkü onun için özel metodlar var. Maksimumu, standart sapması gibi istatistiksel bilgileri kolaylıkla elde edebilmemiz mümkün.

```python
data2.resample("A").mean()
```

data2'yi yıla (A) göre resample ettik. data2'de 1992 ve 1993 şeklinde 2 farklı yıl var. Yani data2'yi yıla göre resample ettiğimizde sonuç olarak 2 tane yıl elde edicez. Resample et ve mean() metodundan dolayı ortalamasını al demiş olduk.

Böylece tüm yıllara göre resample edip, o yılların tüm değerlerinin ortalamasını aldık. 

Çıktıyı inceleyelim. 1992 yılındaki pokemonların can (HP) ortalaması 61.6 ymış. Bunun anlamı, 1992 yılında 3 tane pokemon var ve data2'de görebileceğiniz üzere bunların can (HP) değerleri sırasıyla 45,60,80 dir. (45+60+80)/3=61,6 elde edilir.

```python
data2.resample("M").mean()
```

Aylara göre resample ettik. Bir yılda 12 ay olduğunu biliyoruz ve 1992 yılına baktığımızda 1. , 2. ve 3. aylar için resamble yaptık. Fakat 4.aya geçtiğimizde data'da 4.aydan sonrası olmadığı için bunlara NaN value'si atanıyor.

Ayrıca dikkatinizi çekerim ki, 1992 yılının 1. ayındaki değerlere baktığımızda data2'deki değerler ile aynı. Bunun sebebi, 1992 yılının 1.ayına ait tek bir satır olmasıdır. 

1993 yılının 3.ayına ait 2 adet satır var ve aylara göre resample yaptığımızda bu 2 satırın ortalamaları buradaki çıktıya yansıtılır. 

1993 yılının 4.ayına NaN yazmama sebebi, bu metodda en erken yıl (1992) ile en geç yıl (1993) arasını aylara göre doldurur.

```python
data2.resample("M").first().interpolate("linear")
```

Aylara göre resample edilen data'yı linear bir şekilde interpolate ediyoruz. Bu şu demek:

1,2,3, ,5,6,7

Aradaki boşluğa 4 yazılır. Linear interpolate budur. 

Çıktıda görüldüğü üzere, 1992 yılının 1., 2. ve 3. aylarına sırasıyla 1.000000	, 2.000000, 3.000000 yazmış ve sonrasındaki NaN değerleri 4.000000 'e kadar artan sırada doldurmuş. Yani 3.000000 ile 4.000000 arasında eşit aralıklara bölerek artırmış. 0.08 artırarak doldurmuş.
