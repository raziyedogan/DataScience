TIDY DATA

Verileri pandas kütüphanesinin bir fonksiyonu olan melt() ile düzenleyebiliyoruz.

```python
import pandas as pd

data = pd.read_csv("pokemon.csv")
```

Veri setinin ilk 5 satırını bir değişkene atayalım.

```python
data_new = data.head()
data_new
```

Şimdi de data'yı melt edelim. 

Melt etmek ne demek?: Veri setinden yeni feature'ler çıkartarak veri setini değerleri sabit kalacak şekilde farklı bir yapıya büründürmektir.

```python
melted = pd.melt(frame=data_new,id_vars = 'Name', value_vars= ['Attack','Defense'])
melted
```

Bir üst satırda olanları açıklayalım.

'frame=data_new' parametresi ile yapılan: Ment edeceğimiz data'nın veri setinin ilk 5 satırını tutan data_new olduğunu belirttik.

id_vars = 'Name' parametresi ile yapılan: Veri setinde değişmeden kalmasını istediğimiz feature'nin ismini belirttik. Yani ment işleminden sonra melted değişkeninde tutulan veri seti 'Name' feature'sine sahip olacaktır. Yani her bir pokemonun ismine göre melt ediyoruz.

value_vars= ['Attack','Defense'] parametresi ile yapılan: Yeni oluşmasını istediğimiz değişkenler, Attack ve Defense değerlerinden ortaya çıkacaklar. İlk feature'nin ismi 'variable', ikinci feature'nin ismi 'value' dir. Bu isimleri melt fonksiyonu default olarak atar. İstenirse bu isimlerin değiştirilmesi mümkün.

Çıktıyı inceleyelim.

İlk satırı okuyalım. Name değeri 'Bulbasaur'. Bulbasaur'ın Attack'ı variable feature değeridir. Attack sütunundaki değeride value feature değeridir. Böylelikle Bulbasaur'ın variable sütununa Attack, Bulbasaur'ın value sütununa 49 değeri verilmiştir.
Bu durumda data_new'de tutulan ilk satırı melt etmiş olduk.

İkinci satırı okuyalım. Name değeri Ivysaur. Ivysaur'ın variable'si Attack'tır. data_new'de görüldüğü üzere Ivysaur'ın Attack sütunundaki değeri 62 dir ve bu durumda value'si 62'dir. 

Son satırı okuyalım. Name değeri Charmander. Charmander'ın variable'si Defense. data_new'de görüldüğü üzere Charmander'ın Attack sütunundaki değeri 43 dür ve bu durumda value'si 43'dür.

Sonuç olarak, istediğimiz feature'ler doğrultusunda veri setini şekillendirdik. 

Aslında melt metodu pandas ile seaborn kütüphaneleri arasındaki köprü gibidir. Çünkü seaborn kütüphanesi aracılığıyla data'yı çok güzel tool'lar ile görselleştiricez ve bu görselleştirme işlemi için ilk önce data'yı melt etmemiz gerekmektedir.

PIVOTING DATA

Melt işlemi ile yapılanları eski haline döndürme işlemidir.

```python
melted.pivot(index = 'Name', columns = 'variable',values='value')
```

Bir üst satırda olanları açıklayalım.

index = 'Name' parametresi ile yapılan: Değişmeyen sütunu Name olarak seçmiştik. Bundan dolayı index'e Name atanır.

columns = 'variable' parametresi ile yapılan: melt fonksiyonu ile Attack ve Defense sütunlarını kullanmıştık. Bu sütunlar variable sütunu altında bulunmaktaydı. Dolayısıyla columns'a variable atanır.

values='value' parametresi ile yapılan: melt fonksiyonu ile data_new'de Name'e karşılık gelen Attack ve Defense sütunlarınının değerlerini kullanmıştık. Bu değerler value sütunu altında bulunmaktaydı. Dolayısıyla values'e value atanır.

Ve pivot() ile, melt işlemi ile yapılanları eski haline döndürürüz. Böylece eski data'mızı tekrardan elde ederiz.
