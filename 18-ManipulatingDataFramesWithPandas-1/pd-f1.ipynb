```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("pokemon.csv")
```

INDEXING DATA FRAMES

İndeksi kontrol edebilmemizi sağlayan metodları inceleyelim.

```python
data.head()
```

Yukarıdaki kod ile data'nın indeksini # featuresi yap dedik. Bu kısımdan önceki data.head() sonucundaki çıktıya dikkat edin. Artık # sütununun indeks olduğunu görüyoruz. Öncesinde indeks 0 'dan başlıyordu ama artık 1'den başlıyor.

```python
data["HP"][1]
```

data'nın canının (HP) 1. satırına eriştik. İndeks artık 1'den başladığı için ilk satıra erişmek için 1 yazdık.

```python
data.HP[1]
```

Bu ifade, bir öncesindeki ifade ile aynı işlemi gerçekleştirir. data'nın HP column'ının 1.satırındaki değeri yazdırır.

```python
data.loc[1,["HP"]]
```

data'nın 1.satır HP sütununun bulunduğu location'una eriş demektir.

```python
data[["HP","Attack"]]
```

Data'nın HP ve Attack sütunlarını yazdırır.

SLICING DATA FRAME

bu bir şey ya da iki şey seçmek yerine aralık seçme imkanı tanır.

```python
print(type(data["HP"]))     # series
print(type(data[["HP"]]))   # data frames
```

Eğer tek bir tane [] var ise series, iki tane [[]] var ise data frame 'dir. Series ve data frames arasındaki fark budur.

```python
data.loc[1:10,"HP":"Defense"] 
```

HP'den Defense'e kadar olan sütunların 1'den 10'a kadarki indekslerini aldık.

```python
# Reverse slicing 
data.loc[10:1:-1,"HP":"Defense"] 
```

Bir öncesindeki kod sonucunun tersini aldık.

```python
data.loc[1:10,"Speed":]
```

Burada yaptığımız, Speed sütununu ve bu sütundan sonrasını tamamen almaktır. İlgili sütunların 0 ile 10 arasındaki indekslerindeki değerler çıktıya yansır.

FILTERING DATA FRAMES

```python
boolean = data.HP > 200
data[boolean]
```

data'nın HP değeri 200 den büyük olan kısımlara True atanır. Ve bunu data'nın içerisine koyduğumuzda False'leri yazdırmaz, True'lerin hepsini yazdırır.

```python
# Combining filters
first_filter = data.HP > 150
second_filter = data.Speed > 35
data[first_filter & second_filter]
```

İki adet filtre oluşturduk ve bu iki filtrenin kesişimi olanlar çıktıya yansır. Yani HP değeri 150'den büyük olanlar ve aynı zamanda Speed değeri 35 'den büyük olanlar çıktıda görülür.

```python
data[data.Speed<15]
```

Filtremiz Speed değerinin 15'ten küçük olduğu durumlardır. Yukarıda bu durumu sağlayan satırları görüyoruz.

```python
# Filtering column based others
data.HP[data.Speed<15]
```

Filtremiz Speed değerinin 15'ten küçük olduğu durumlardır. Bu durumları sağlayan satırların HP değerlerini görüntüledik.

TRANSFORMING DATA

```python
def div(n):
    return n/2
data.HP.apply(div)
```

div isimli bir metod oluşturduk ve bu metod bir değer alarak, bu değeri 2'ye bölme işlemi yapıyor. 

"data.HP.apply(div)" satırı ile, Pokemonların HP değerlerini 2'ye bölüyoruz. 

Veya lambda fonksiyonunu kullanabiliriz.

```python
data.HP.apply(lambda n : n/2)
```

Lambda fonksiyonu, metod yazmanın kısa yoludur. 

(lambda n : n/2) kısmı ile ifade edilen; bir n inputu olsun ve n/2 'yi return etsin.

```python
data["total_power"] = data.Attack + data.Defense
data.head()
```

Yeni bir "total_power" isminde column tanımladık. Bu sütunun değerleri Attack ve Defense sütun değerlerinin toplamıdır.
