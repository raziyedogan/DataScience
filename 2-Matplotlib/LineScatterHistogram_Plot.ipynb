Matplotlib kütüphanesi, görselleştirme konusunda bize yardım eder. Line, scatter ve histogram en temel plot'lardır.

Line plot'u ne zaman kullanırız? : Eğer x ekseni 'zaman' ise line plot kullanarak görselleştirme yapabiliriz.

Scatter plot'u ne zaman kullanırız? : İki tane feature arasında correlation olup olmadığını anlamak için scatter plot kullanabiliriz.

Histogram plot'u ne zaman kullanırız? : Eğer bir data'nın mesela bir feature'nin sıklığını anlamak istiyorsak histogram plot kullanabiliriz. 
feature'nin sıklığı ne demek?: Mesela bir sütunda 5 adet aynı değerden var ise bu sıklığı ifade eder. Histogramı bu durumu görselleştirmek için kullanabiliriz.

Haydi bu anlatılanları koda dökelim.

```python
#Tabiki her şeyden önce kütüphaneleri çağıralım.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('pokemon.csv')
```

```python
#LINE PLOT

data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

Bu kod bloğunda yaptıklarımızı adım adım açıklayalım:

data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':') satırı: 

Bu satırda, Speed column'ını aldık ve bunu plot ettirdik. kind parametresine line, scatter veya histogram yazabiliriz. Burada line plot kullandık. color değerine yeşil verdik. label değerine 'Speed' yazarak yeşil renk ile gördüklerimizin Speed'i ifade etmesini sağladık. linewidth parametresi line plot'un kalınlığını ifade eder. alpha parametresi saydamlığı ifade eder. grid parametresi ile arka plandaki karelerin var olmasını sağladık ve böylece line'lara karşılık gelen sayısal değerleri arka plandaki çizgileri takip ederek rahatlıkla görebiliriz. linestyle ile line'ların  nasıl çizileceğini ayarladık.

data.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.') satırı:

Bu satırda, data'mızın Defense column'ını belirlediğimiz parametrelere göre plot ettirdik.

Çıktıda x ekseninde yer alanlar pokemonları ifade eder. y ekseni ise pokemonların hızı (Speed) ve defans gücüdür (Defense).

```python
#SCATTER PLOT

# x = attack, y = defense
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')   
```

Amacımız iki tane feature arasındaki ilişkiyi yani correlation'u bulmak ise Scatter plot güzel bir tercih. 

attack ve defense arasındaki ilişkiyi inceledik. Attack'ı x eksenine, Defense'yi y eksenine koyduk. Çıktıdan gördüğümüz üzere sezgisel olarak 'genelde Attack artıyorsa Defense'de artıyor' diyebiliriz. 

data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red') satırında:

kind parametresi ile scatter plot olacağını belirttik. x ekseninde data'nın 'Attack' ı olacağını ve y ekseninde data'nın 'Defense' 'i olacak. Dikkat edilmesi gerkeen bir nokta, x ve y değerlerine atanan ifadeler column isimlerinin aynısı olmalıdır. Mesela Attack yerine ilk harfi küçük olacak şekilde attack yazarsak hata meydana gelir. 

Ayrıca bu çizdirme işlemini 'plt.scatter(data.Attack, data.Defense, color="red", alpha=0.5)' şeklindede yapabilirdik. Yukarıda yaptığımız ile aynı sonucu verir.

plt.xlabel('Attack') satırı: x ekseninin ismini belirledik.
plt.ylabel('Defence') satırı: y ekseninin ismini belirledik.
plt.title('Attack Defense Scatter Plot')  satırı: Başlığı belirledik.

Böylelikle scatter plot'u çizdirdik ve çıktıdan gördüğümüz üzere, defense ile attack doğru orantılı olarak artmaktadır.

```python
#HISTOGRAM

data.Speed.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()
```

data.Speed.plot(kind = 'hist',bins = 50,figsize = (12,12)) satırı:

data'nın Speed'ini plot ettirdik. kind parametresi ile histogram plot olmasını sağladık. Çıktıda görünen çubuklara bar denir ve bins parametresi ile barların sayısını ayarlarız. figsize parametresi ile çıktımızdaki frame'in boyutunu ayarlarız.

x ekseni data'nın yani pokemonların sahip olduğu hız (Speed) değerleridir. y ekseni ise frekanslarıdır yani x eksenindeki hız değerinden kaç adet olduğu bilgisidir.

Mesela output'u okuyalım: 50 hızına sahip pokemonların sayısı 56 civarındadır.

Histogram'dan elde ettiğimiz bilgiler istatistiksel sonuçlar elde etmek açısından önemlidir.

```python
#clf() metodu
data.Speed.plot(kind = 'hist',bins = 50)
plt.clf()
```

clf() metodu ile çizdiğimiz plot'u yok edebiliriz. Başka bir deyişle clear edebiliriz.
