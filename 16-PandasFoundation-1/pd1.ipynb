BUILDING DATA FRAMES FROM SCRATCH

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data frames from dictionary
country = ["Spain","France"]
population = ["11","12"]
list_label = ["country","population"]
list_col = [country,population]
zipped = list(zip(list_label,list_col))
data_dict = dict(zipped)
df = pd.DataFrame(data_dict)
df
```

Yukarıda yapılanları açıklayalım.

Spain ve France değerlerinden oluşan country isimli liste oluşturduk.

11 ve 12 değerlerini barındıran population isimli liste oluşturduk.

country ve population'u list_col isimli liste içerisine attık.

Sonrasında list_label ile list_col 'u zipledik yani sıkıştırdık. Bunun sonucunda zipped isimli bir file oluşturmuş olduk.

Sonra zipped file'ını dictionary'e çevirdik. 

Sonrasında da dictionary'den dataframe'i elde ettik.

Böylece öncelikle dictionary oluşturduk ve daha sonra dataframe oluşturduk.

Şimdi oluşturduğumuz dataframe'e yeni bir sütun ekleyelim.

```python
#Yeni bir column:

df["capital"] = ["madrid","paris"]
df
```

Daha öncesinde oluşturduğumuz dataframe'e capital isimli bir sütun ekledik. Sütuna madrid ve paris değerlerini ekledik.

Broadcasting: Yeni bir sütun oluşturularak, bu oluşturulan sütunun tamamına aynı değerin atanmasıdır.

Hadi uygulayalım.

```python
# Broadcasting
df["income"] = 0 #Broadcasting entire column
df
```

dataframe'e income isimli bir sütun ekledik ve bu sütuna 0 değerini atadık.

VISUAL EXPLORATORY DATA ANALYSIS

Görselleştirme denince akla Plot, Subplot ve Histogram gelir.

Plot ile başlayalım. Ama tabiki öncesinde veri setimizi getirelim.

```python
data = pd.read_csv("pokemon.csv")
```

```python
data1=data   
data1["Type 2"].dropna(inplace = True)
```

```python
# Plotting all data 
data1 = data.loc[:,["Attack","Defense","Speed"]]
data1.plot()
```

Pokemon veri setinin Attack, Defense, Speed bilgilerini tek bir plot üzerinde çizdirdik.

```python
# subplots
data1.plot(subplots = True)
plt.show()
```

Attack bir subplot'ta, Defense başka bir sublot'ta ve Speed başka bir sublot'tadır.

```python
# scatter plot  
data1.plot(kind = "scatter",x="Attack",y = "Defense")
plt.show()
```

x ekseni Attack'ı ve y ekseni Defense'ı ifade eder. Ve bu iki feature arasındaki ilişkiyi yani korelasyonu görüntülemek amacıyla scatter plot kullanabiliriz.

```python
# hist plot  
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250))
```

Histogram, frekans ölçer.

range parametresi y eksenidir ve 0 'dan 250'ye kadar olması belirtilmiş.

```python
# histogram subplot with non cumulative and cumulative
fig, axes = plt.subplots(nrows=2,ncols=1)
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250),ax = axes[0])
data1.plot(kind = "hist",y = "Defense",bins = 50,range= (0,250),ax = axes[1],cumulative = True)
plt.savefig('graph.png')
plt
plt.show()  #Bu satır yazılmadığında çıktının en üstünde bir yazı görünür. Bu yazının görünmesini istemiyorsak bu satırı ekleriz.
```

Çıktıda gördüğümüz histogramdır. 

Kodda gördüğünüz 'cumulative = True' parametresi şudur: Mesela 50 'yi çizdirirken, 50 değerine karşılık gelen kısım kendinden önceki kısmın toplamıdır. Bu yüzden çıktıda alttaki görselde gördüğünüz şekil artarak gidiyor.
