```python
f,ax = plt.subplots(figsize=(14,14))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
plt.show()   #çıktıda üst tarafta bir yazı bulunuyor. Bu yazının görünmesini istemezsek bu satırı ekleriz.
```

Bir üst satırda neler olduğunu açıklayalım.

Correlation map: Feature'ler arasındaki ilişkiyi anlamamızı sağlar. Eğer 2 tane feature arasında correlation 1 ise, bunlar birbiri ile doğru orantılıdır. Mesela bir evin oda sayısı artar ise fiyatıda artar. Bu durumda bir evin oda sayısı ile fiyatı pozitif ilişkilidir. Yani positive correlation'dir. Mesela bir evin şehir merkezine olan uzaklığı artarsa, evin fiyatı azalır. Bu durumda evin şehir merkezine olan uzaklığı ile fiyatı ters orantılıdır. Yani negative correlation'dir. Feature'ler arasındaki ilişkiyi bulabilmemiz için seaborn kütüphanesinde bulunan heatmap isimli görselleştirme fonkisyonunu kullanacağız. Fonksiyona parametre olarak 'data.corr()' veriyoruz. data.corr(), feature'ler arasındaki correlation'u yani ilişkiyi verir. Her bir sütun'un diğer sütunlarla arasındaki ilişki sayısal bir değer ile yansıtılır. Dikkat etmek gerek ki, bir sütunun kendisi ile arasındaki ilişkinin değeri 1 dir. Çünkü kıyaslanan sütunlar aynı olduğundan biri azalırsa diğeride azalacaktır.

Eğer değer 1 veya 1'e yakın ise positive correlation yani doğru orantılıdır. -1 veya -1'e yakın bir değer ise, negative correlation yani ters orantılıdır. Eğer 0 ise, sütunlar arasında bir ilişki yoktur denilebilir. Mesela bir uçak bileti fiyatı ile bir kütüphanedeki kitap sayısının hiçbir alakası yok. Bu iki feature birbiriyle hiçbir alakası olmayan şeylerdir. Bu yüzden bu iki feature arasındaki correlation 0'dır.

heatmap fonksiyonu sayesinde data.corr() dan elde edilen tabloyu daha güzel,anlaşılır bir görsel olarak görebiliyoruz.

heatmap fonksiyonunun parametreleri şu şekildedir:

1.parametre: data.corr()

2.parametre: 'annot=True' ile çıktıda görsel üzerindeki sayıların görünmesi sağlanır.

3.parametre: 'linewidths=.5', çıktıdaki görselde her bir hücre arasındaki çizginin kalınlığını ifade eder.

4.parametre: 'fmt='.1f', çıktıda hücrelerde yer alan değerin noktadan sonraki hane sayısını ifade eder.

5.parametre: 'ax=ax' ifadesinde yer alan ax değişkeni bir üst satırdadır. 'f,ax = plt.subplots(figsize=(18,18))' satır ile şeklin boyutu belirlenmiştir. Bu satırı yazmasakta olurdu fakat bu sefer default bir değer ile boyut ayarlanacağından çıktıda küçük boyutlu bir şekil elde edilebilirdi.

Elde ettiğimiz çıktıyı inceleyelim: 

Mesela Area ile Perimeter arasında 0.9 oranında correlation var. Renginden de anlaşılacağı üzere bu mükemmel. Bu feature'ler arasındaki correlation 1'e oldukça yakın olduğu için positive correlation yani doğru orantı söz konusudur.

Mesela Compactness ile Eccentricity arasında -1.0 oranında correlation var. Renginden de anlaşılacağı üzere bu berbat. Bu feature'ler arasındaki correlation -1 olduğu için negative correlation yani ters orantı söz konusudur. 



Şimdi biraz Matplotlib kütüphanesinden faydalanalım.

Matplotlib kütüphanesi, görselleştirme konusunda bize yardım eder. Line, scatter ve histogram en temel plot'lardır.

Line plot'u ne zaman kullanırız? : Eğer x ekseni 'zaman' ise line plot kullanarak görselleştirme yapabiliriz.

Scatter plot'u ne zaman kullanırız? : İki tane feature arasında correlation olup olmadığını anlamak için scatter plot kullanabiliriz.

Histogram plot'u ne zaman kullanırız? : Eğer bir data'nın mesela bir feature'nin sıklığını anlamak istiyorsak histogram plot kullanabiliriz. feature'nin sıklığı ne demek?: Mesela bir sütunda 5 adet aynı değerden var ise bu sıklığı ifade eder. Histogramı bu durumu görselleştirmek için kullanabiliriz.

Haydi bu anlatılanları koda dökelim.

```python
#LINE PLOT

data.Area.plot(kind = 'line', color = 'g',label = 'Area',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Perimeter.plot(color = 'r',label = 'Perimeter',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

Bu kod bloğunda yaptıklarımızı adım adım açıklayalım:

data.Area.plot(kind = 'line', color = 'g',label = 'Area',linewidth=1,alpha = 0.5,grid = True,linestyle = ':') satırı:

Bu satırda, Area column'ını aldık ve bunu plot ettirdik. kind parametresine line, scatter veya histogram yazabiliriz. Burada line plot kullandık. color değerine yeşil verdik. label değerine 'Area' yazarak yeşil renk ile gördüklerimizin Area'yı ifade etmesini sağladık. linewidth parametresi line plot'un kalınlığını ifade eder. alpha parametresi saydamlığı ifade eder. grid parametresi ile arka plandaki karelerin var olmasını sağladık ve böylece line'lara karşılık gelen sayısal değerleri arka plandaki çizgileri takip ederek rahatlıkla görebiliriz. linestyle ile line'ların nasıl çizileceğini ayarladık.

data.Perimeter.plot(color = 'r',label = 'Perimeter',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.') satırı:

Bu satırda, data'mızın Perimeter column'ını belirlediğimiz parametrelere göre plot ettirdik.

Çıktıda x ekseninde yer alanlar kabak çekirdeklerini ifade eder. y ekseni ise kabak çekirdeklerinin alan ve çevreleridir.

```python
#SCATTER PLOT

data.plot(kind='scatter', x='Convex_Area', y='Major_Axis_Length',alpha = 0.5,color = 'blue')
plt.xlabel('Convex_Area')              
plt.ylabel('Major_Axis_Length')
plt.title('Convex_Area Major_Axis_Length Scatter Plot')  
```

Amacımız iki tane feature arasındaki ilişkiyi yani correlation'u bulmak ise Scatter plot güzel bir tercih.

Convex_Area ve Major_Axis_Length arasındaki ilişkiyi inceledik. Convex_Area'ı x eksenine, Major_Axis_Length'yi y eksenine koyduk. Çıktıdan gördüğümüz üzere sezgisel olarak 'genelde Convex_Area artıyorsa Major_Axis_Length'de artıyor' diyebiliriz.

data.plot(kind='scatter', x='Convex_Area', y='Major_Axis_Length',alpha = 0.5,color = 'blue') satırında:

kind parametresi ile scatter plot olacağını belirttik. x ekseninde data'nın 'Convex_Area' ı olacağını ve y ekseninde data'nın 'Major_Axis_Length' 'i olacak. 

plt.xlabel('Convex_Area') satırı: x ekseninin ismini belirledik. 

plt.ylabel('Major_Axis_Length') satırı: y ekseninin ismini belirledik. 

plt.title('Convex_Area Major_Axis_Length Scatter Plot') satırı: Başlığı belirledik.

Böylelikle scatter plot'u çizdirdik ve çıktıdan gördüğümüz üzere, Convex_Area ile Major_Axis_Length doğru orantılı olarak artmaktadır.
