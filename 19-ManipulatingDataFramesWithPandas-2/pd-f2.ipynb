PIVOTING DATA FRAMES

Feature column olarak kullandığımız isimleri artık birer value ile yapıcaz demek. Yani mesela Attack ile Attack sütunun 1.satırındaki 49 değerinin yer değiştirmesi gibi. Artık 49 feature name olucak ve Attack 'da bir değer olucak.

Bunu uygulamak için bir data frame oluşturalım.

```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df = pd.DataFrame(dic)
df
```

Oluşturduğumuz data frame'in kurgusu şu şekildedir:

treatment: Tedavidir.

gender: Cinsiyettir.

response: Tedaviye yanıt verebilme durumudur.

age: İnsanın yaşıdır.

```python
# pivoting
df.pivot(index="treatment",columns = "gender",values="response")
```

Yeni indeksi "treatment" yaptık. İndeksler A ve B oldu. 

Column'lar yani feature name'ler gender'deki değerler oldu. Yani F (kadın) ve M (erkek) ŞEKLİNDE iki çeşit cinsiyet vardı ve yeni feature isimleri F ve M oldu.

Bu sütunların değerleride response sütunundaki değerler oldu.

STACKING and UNSTACKING DATAFRAME

```python
df1 = df.set_index(["treatment","gender"])
df1
```

treatment'ın outer ve gender'ın inner indeks olmasını sağladık. Yani ilk başta treatment'a göre diziyor ve sonra gender'a göre diziyor.

```python
df1.unstack(level=0)
```

unstack() metodu, birden fazla indekse sahip data frame'lerin indeklerinin bir tanesinden kurtulmamızı sağlar.

treatment, indeks 0'dır. Ve gender, indeks 1'dir.

Burada indeks 0 'ı unstack yaptık. Yani treatment artık indeks değildir. Sadece gender indekstir.

```python
df1.unstack(level=1)
```

```python
df2 = df1.swaplevel(0,1)
df2
```

swaplevel metodu ile indekslerin yerlerini değiştirmemiz mümkün.

Yani variable sütun değerine df'deki sütun yazılır ve value sütununada df'deki variable sütununa yazılan sütuna karşılık gelen değer yazılır.

CATEGORICALS AND GROUPBY

```python
df
```

```python
df.groupby("treatment").mean()
```

groupby() metodu ile bir şeye göre gruplama yapılır. Burada treatment'a göre gruplama yapılıp, ortalaması alınmıştır.

Çıktıyı inceleyelim. treatment'ı A olan response değerleri 10 ve 45'dir. (10+45)/2=27.5'dir. Görüldüğü üzere treatment'ı A ve sütunu response olan hücrede 27.5 yazmaktadır.

```python
df.groupby("treatment").age.mean() 
```

Burada yapılan, data frame treatment'e göre gruplanır ve sonra bunun sadece age'i alınır ve bunun ortalaması alınır.

Çıktıyı inceleyelim. Treatment'i A olanların age değerleri 15 ve 4'tür. (15+4)/2=9.5'tir. 

```python
df.groupby("treatment").age.max() 
```

Burada yapılan, data frame treatment'e göre gruplanır ve sonra her bir grubun age değeri maksimum olan alınır.

Çıktıyı inceleyelim. treatment'i A olan age değerleri 15 ve 4 'tür. Bunlardan en büyüğü olan 15 değeri çıktıya yansıtılır.

```python
df.groupby("treatment")[["age","response"]].min()
```

Birden fazla feature var. Bu feature'leri treatment'a göre grupladık ve sonra içerisinden sadece age ve response sütunlarını alıp min() metodunu çağırdık.

Yani treatment'e göre grupladık ve daha sonra age ve response için minimum değerlerini bulduk.
